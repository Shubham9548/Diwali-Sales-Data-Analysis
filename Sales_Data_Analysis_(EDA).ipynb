# Exploratory Data Analysis

Gender

```python
df.columns
```

```python
# Count of Orders
gen=sns.countplot(x='Gender',data=df)
for i in gen.containers:
    gen.bar_label(i)
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
sns.barplot(x='Gender',y='Amount',data = sales_gen)
```

from above graphs we can see that most of the buyers are females and even the purchasing power of females are greater then men

# Age

```python
df.head(2)
```

```python
age=sns.countplot(x='Age Group',data=df)
for i in age.containers:
    age.bar_label(i)
```

```python
age=sns.countplot(x='Age Group',hue='Gender',data=df)
for i in age.containers:
    age.bar_label(i)
```

```python
sales_age=df.groupby(['Age Group','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x='Age Group',y='Amount',hue='Gender',data=sales_age)
plt.show()
```

 from above graph we can see that most of the buyers are age group between 26-35yrs female

# State

```python
df.head()
```

```python
sales_state=df.groupby(['State'], as_index=False)["Orders"].sum().sort_values(by='Orders', ascending=False)
```

```python
sales_state
```

```python
sns.barplot(x='State',y='Orders',data=sales_state)
plt.xticks(rotation=90)
plt.show()
```

```python
state_amou=df.groupby(['State'], as_index=False)["Amount"].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x='State',y='Amount',data=state_amou)
plt.xticks(rotation=90)
plt.show()
```

From above graphs we can see that most of the orders & total sales/amount are from UttrarPradesh,Maharashtra and Karnataka respectively

# Marital Status

```python
df.head(1)
```

```python
ms=sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize':(7,5)})
for i in ms.containers:
    ms.bar_label(i)
```

```python
sales_ms=df.groupby(['Marital_Status','Gender'], as_index=False)["Amount"].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x='Marital_Status',y="Amount",hue='Gender',data=sales_ms)
plt.show()
```

from above graph we can see that most of the buyers are married (women) and the have high purchasing power

# Occupation

```python
df.head(1)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
occ=sns.countplot(x='Occupation',data=df)

for i in occ.containers:
    occ.bar_label(i)
```

```python
sales_occ=df.groupby(['Occupation'], as_index=False)["Amount"].sum().sort_values(by='Amount', ascending=False)
```

```python

sns.barplot(x="Occupation",y="Amount",data=sales_occ)
plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

# Product Category

```python
df.head(1)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
pro=sns.countplot(x="Product_Category",data=df)
plt.xticks(rotation=75)

for i in pro.containers:
    pro.bar_label(i)
```

```python
sales_pro=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x='Product_Category',y='Amount', data=sales_pro)
plt.xticks(rotation=75)
plt.show()
```

 from above graphs we can see that most of the sold product are from Food, Clothing and Electronics category
